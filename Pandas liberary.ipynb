```python
#imporing library for the data integration
import pandas as pd
```

```python
#Data to series from

marks=[88,99,77,60,69]
std_marks=pd.Series(marks)
print(std_marks)
```

```python
#convertind data to series
marks=pd.Series([88,99,77,60,69],index=["Math","Science","Engl","Hindi","Geo"])
print(marks)
```

```python
#Accessing series by index value
print(marks[0:4])
```

```python
#Updating series by valus
marks['Hindi']=100
print(marks)
```

```python
#updating series by index numbers
marks[2]=90
print(marks)
```

```python
#performing some arithmatic functions(/,+,-,*)

# ADD 
upd_marks=pd.Series([88,99,77,60,69])
print(upd_marks + 5)
```

```python
# SUBTRACT
upd_marks=pd.Series([88,99,77,60,69])
print(upd_marks - 5)
```

```python
# MULTIPLY
upd_marks=pd.Series([88,99,77,60,69])
print(upd_marks *5)
```

```python
#Divide
upd_marks=pd.Series([88,99,77,60,69])
print(upd_marks /5)
```

```python
#Identifying missing value in dataset

missing_marks=pd.Series([88,None,77,None,69])
print(missing_marks)
```

```python
#handling missing value
missing_marks.fillna(0,inplace=True)
print(missing_marks)
```

```python
#handling missing value
missing_marks.fillna(1,inplace=True)
print(missing_marks)
```

```python
data={
    "Name":["Amit","Ajeet","Raj","Vivek","Saurabh"],
    "Marks":[85,None,70,88,90]
}
df=pd.DataFrame(data)
```

```python
#check null values
df.isnull()
```

```python
#Fill null by specify value
df.fillna(70)
```

```python
#remove null value
df.dropna()
```

```python
#filll missing valuye t=with average of other values
df["Marks"].fillna(df["Marks"].mean(),inplace=True)
```

```python
data={
    "Trade":["AIPA","AIPA","Cosmo","CSA","COPA","Cosmo","CSA"],
    "Students":[20,20,40,35,30,20,40]
}
df=pd.DataFrame(data)
```

```python
df
```

```python
df.groupby("Trade")["Students"].sum()
```

```python
df.groupby("Trade")["Students"].mean()
```

```python
df.groupby("Trade")["Students"].count()
```

```python
#used for summerizing data
df.groupby("Trade")["Students"].agg(["sum","mean","count"])
```
